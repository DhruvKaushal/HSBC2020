## Functions

```python
def check_prime(num):
    '''
    This is the docstring for this function.
    This function takes a number (int) and checks if it is prime.
    The function return True / False (bool)

    '''
    for x in range(2,int(num/2)+1):
#         print(x)
        if num%x == 0:
            return False
    else:
        return True
```

```python
def adder(*args):
    res = 0
#     print(args)
    for x in args:
        res = res + x
    return res
```

```python
adder(5,6,7,8,9,10)
```

```python
def adder(lst1):
    res = 0
#     print(args)
    for x in lst1:
        res = res + x
    return res
```

```python
adder(nums)
```

```python
def square(num1, num2):
    return num1*num2
```
